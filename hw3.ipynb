# 1. Leibniz Pi formula


$$\begin{align}
\frac{\pi}{4} = \frac{1}{1}-\frac{1}{3}+\frac{1}{5} -\frac{1}{7} + \frac{1}{9} - \cdots 
\end{align}$$
이 식은 라이프니츠의 파이 공식이다.

1. 다음의 식을 구현하는 함수를 만들어라.
$$\begin{align}
f(n) = \frac{1}{1}-\frac{1}{3}+\frac{1}{5} -\frac{1}{7} + \frac{1}{9} - \cdots + \frac{1}{n}.
\end{align}$$
2. 이 식을 이용하여 아래처럼 그림을 그려라.

```python
def f(x):
    if x == 0 :
        return(0)
    elif x > 0 :
        return((-1)**(x-1)*1/(2*x-1))
def Leibniz(x): 
    a=[]
    for i in range(x):
        i += 1
        a.append(f(i))
    return(np.sum(a))
```

```python
A=[x for x in range(1,100)] 
B=[Leibniz(x) for x in range(1,100)] 
plt.plot(A,B,color='blue', linewidth=0.7, label='Leibniz formula for $\pi$',) 
plt.axhline(y=np.pi/4,color='orange', linewidth=0.7,label='$\pi$/4')
plt.xlabel('n')
plt.legend()
```

# 2. Logistic Map


로지스틱 맵은 다음처럼 주어진다.$$ f_{n+1} = r f_n (1- f_n). $$


작성한 함수를 이용하여 다음의 조건에서 $n=1$에서 $100$까지 로지스틱 맵의 값을 구하고 그래프로 그려라. 즉, x축은 $n$, y축은 $f_n$인 그래프를 그려라.

$r=0.5$, $f_0=0.5$

$r=0.95$, $f_0=0.5$

$r=2$, $f_0=0.5$

$r=3.2$, $f_0=0.5$

$r=3.8$, $f_0=0.5$

$r=3.8$, $f_0=0.501$

$f_n$을 구하는 함수를 작성하여라.

$r=0.5$, $f_0=0.5$

```python
def logistic(r, f, n):
    for i in range(n):
        f = r*f*(1-f)
    return f
```

```python
for x in range(1,101) :
    print(x,logistic(0.5,0.5,x))
```

```python
a=[x for x in range(1,101)]
b=[logistic(0.5,0.5,x) for x in range (1,101)]
plt.plot(a,b,color='blue',label='r = 0.5, $f_0$ = 0.5') 
plt.xlabel('n')
plt.ylabel('f_n')
plt.legend()
```

# 3. sinc function


1. 싱크 함수를 그려보아라. 싱크함수는 다음처럼 주어진다.$$ sinc(x)  = \frac{\sin x}{x}. $$


```python
x=np.arange(-10,10,0.1)
def sinc(x):
    return(np.sin(x)/x)
plt.plot(x,sinc(x))
```

2. 싱크 함수를 테일러 전개하면 다음과 같다.$$ sinc(x)  = \frac{sin x}{x} = \sum_{n=0}^{\infty} \frac{(-1)^n x^{2n} }{(2n+1)!}. $$위의 식의 우변을 $n$을 바꾸어 가면서 그래프로 그리고 싱크함수와 비교하여라.

```python
import math
```

```python
def f(x,N):
    i, s = 0, 1
    while (i<=N):
        s = s + ((-1)**i*x**(2*i))/(math.factorial(2*i+1))
        i = i+1
    return s
```

```python
x=np.arange(-10,10,0.1)
plt.plot(x,f(x,1),color='purple')
plt.plot(x,f(x,3),color='red')
plt.plot(x,f(x,10),color='orange')
plt.plot(x,f(x,70),color='blue')
plt.plot(x,f(x,84),color='green') # N=84 까지만 계산 가능. 점점 비슷한 그래프 꼴로 변하고 있음.
plt.ylim(-0.3,2.2)
```
